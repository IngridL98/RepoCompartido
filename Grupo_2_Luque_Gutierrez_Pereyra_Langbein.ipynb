```python
#Importamos las librerias con las funciones que utilizaremos
from sympy import symbols, Function, Derivative,Eq ,plot,plot_implicit, solve, dsolve, simplify, exp, sqrt,pi, cos, simplify,sin,evaluate,And, diff
```

```python
#Declaramos las funciones y las variables simbólicas

t = symbols('t')  #define la variable independiente que será el tiempo
x = Function('x') #define una función movimiento respecto del tiempo
M = symbols('M')  #define la variable correspondiente a la masa
K = symbols('K')  #define la variable correspondiente a la constante de elasticidad
C = symbols('C')  #define la variable correspondiente a la constante de amortiguamiento
```

```python
#Definimos la EDO homogenea del sistema
EDO = Eq( M * Derivative(x(t),t,t) + C * Derivative(x(t),t) + K * x(t),0)
```

```python
#Resolvemos la Ecuacion Diferencial
EDO_sol = dsolve(EDO)
```

## Análisis del confort y desplazamiento sobre una superficie

<p align="justify">Uno de los requerimientos del amortiguamiento que se nos ha solicitado para el traslado es poder garantizar un grado de confort para la carga, es por esto que tomaremos como referencia estudio que determina que el confort se alcanza cuando la <b>frecuencia fundamental (de la oscilación libre amortiguada) del sistema (w)</b> se encuentra entre los 1 − 3 Hz y la <b>relación de amortiguamiento del sistema (E)</b> está entre los 0,24 y 0,26. Tener presente que esta relación está dada por la constante de amortiguamiento del sistema y la constante de amortiguamiento crítica del mismo <b>E = C/Cc</b>.

<p align="justify">Habiendo ya calculado <b>w</b> anteriormente, ahora nos falta obtener una expresión para <b>E</b> y para ello debemos calcular el coeficiente de amortiguamiento crítico del sistema, es decir el valor de <b>C</b> que convierte a nuestro sistema en uno críticamente amortiguado, y algebraicamente lo obtenemos al hallar la expresión para dicho coeficiente que logra hacer 0 el discriminante en la obtención de las raíces.

<p align="justify">El hecho de que dicho discriminante se anule, hará que las raíces obtenidas sean reales y repetidas y esto terminará afectando con un factor <b>t</b> en una de nuestras soluciones linealmente independientes, provocando así el máximo amortiguamiento del sistema. Si bien en nuestra situación no es deseado ni nos interesa que ello ocurra, obtener este valor <b>Cc</b> será necesario para armar la relación de amortiguamiento que nos influye en la determinación del confort.

<p align="justify">Usando el software obtenemos:

```python
Cc = symbols('Cc') # Definimos la constante de amortiguamiento crítica
Ec_discr_igual_cero=Eq (C**2 - 4 *M*K, 0) # Igualamos el discriminante a cero
Cc= solve(Ec_discr_igual_cero,C).pop(1) #Encontramos el valor de Cc que satisface la ec.
Cc
```

```python
E = symbols('E') # definimos la constante
E = C / Cc  # calculamos la relacion de amortiguamiento del sistema
E
```

Por lo tanto este movimiento deberá ser considerado ahora en un nuevo modelo.

<p align="justify">Por su parte se nos ha dicho que este tramo será la peor condición de la carretera y que por tanto será importante lograr que las autopartes que elijamos puedan cumplir con los requerimientos de amortiguación allí. Este tramo presenta una superficie ondulada con una longitud de onda de <b>6 metros</b> y una amplitud de ondulación de la superficie de <b>0,035 m</b>, la cual se recorrerá a una velocidad de <b>110 km/h</b>.

<p align="justify">Con esta información es posible para nosotros representar la carretera mediante una función con comportamiento sinusoidal y en nuestro caso elegiremos la función <b>seno</b> ya que consideraremos que el comienzo de las irregularidades del camino se producirá en la parte media de la onda (valor 0 en nuestro sistema de referencia).

<p align="justify">Entonces sabiendo que dicha función será de la forma <b>A sen (<i>f</i> m )</b> , en donde <b>A</b> es la amplitud (0,035) y  <i><b>f</b></i> será la frecuencia de la función trigonométrica (2π / 6) que nos permitirá lograr la longitud de onda deseada <b>(sin considerar la velocidad aún)</b>, representaremos la posición de la carretera respecto al equilibrio en función de los metros (<b>m</b>) mediante el software de la siguiente forma:


```python
A = symbols('A')  #define la variable correspondiente a la amplitud
f = symbols('f')  #define la variable correspondiente a la frecuencia
m = symbols('m')  #define la variable independiente metros
A = 0.035 # asignamos el dato de la amplitud
f = ((2*pi)/6) # calculo de la frecuencia que nos da una longitud de onda de 6 metros
carretera = A * sin(f * m) #armamos la función
display(carretera)
plot(carretera, (m, 0, 20),ylabel = 'posición (m)' , xlabel ='longitud (m)')
# Graficamos la funcion carreteara en el intervalo [0,20] metros.
```

```python
#Como podemos comprobrar a los 6 metros la carrertera habrá completado el periodo volviendo a la posición de
#equilibrio 0, es decir su longitud de onda es de 6 metros.

carretera_6m = carretera.subs(m, 6)
print(carretera_6m)
```

<p align="justify">Si consideremos al sistema como lo habíamos modelizada anteriormente mediante la ecuación diferencial de segundo orden <b>M * x''(t) + C* x'(t) + K* x(t) = 0</b> que obtuvimos a partir de la segunda ley de Newton y las fuerzas intervinientes en la situación y ahora agregamos un nuevo movimiento <b>v(t)</b> que será impartido a la masa (como se visualiza en la siguiente figura), este se acoplará  al que teníamos previamente afectando por tanto al comportamiento de cada parte del sistema de amortiguamiento.

<p align="justify">En consecuencia  se llega a la ecuación <b>M * x''(t) + C* ( x'(t) - v'(t) ) + K*( x(t) - v(t) ) = 0</b> y despejando los términos hallamos una expresión no homogénea de la forma <b> M * x''(t) + C* x'(t)  + K* x(t) = C* v'(t) + K*v(t) </b>.

image.png

<p align="justify"> Y habiendo ya modelado una función representativa de la carretera será ésta la que tendremos en cuenta para armar nuestra <b> v(t)</b> pero <b> modificandole la frecuencia para lograr la velocidad de 30,55 m/s</b> que será la que tendrá el movil al recorrerla. Por tanto la función que representa el movimiento impartido en función del tiempo por la carretera sobre nuestro vehículo será:
</br><b>v(t)= 0,035 * sen ( ( (2 pi * 30.55)) / 6 ) * t ) </b>.

```python
#Definimos la función V(t)
v = symbols('v')
v = A * sin(30.55*f * t)
display(v)
plot(v, (t, 0, 1),ylabel = 'posición (m)' , xlabel ='tiempo (s)')

# Graficamos la funcion en el intervalo [0,1] y se puede observar que ahora la longitud de 6 metros
# se recorrerá 30.55/6 veces en un segundo, es decir cumple con la velocidad indicada.
```

```python
v_dif= diff(v, t)

#Definimos la EDO no homogenea del sistema
EDO_no_homogenea = Eq(C*Derivative(x(t), t) + K*x(t) + M*Derivative(x(t), (t, 2)),C*v_dif + K*v)
display(EDO_no_homogenea)
```

<p align="justify">Entonces estableciendo que la posición inicial <b>x(to)=Xo</b> será con el tiempo cero y este será el tiempo en donde las ruedas del móvil hagan contacto con la carretera modelada, en ese instante por la función elegida estaremos a la altura <b>0</b> y estableceremos que nuestro elemento elástico no estará estirado, entonces tenemos <b>x(0)= 0</b> . Por otro lado fue dada como dato la velocidad a la cual se recorre el tramo y es de <b>110 km/h</b> , por tanto haciendo la conversión al sistema métrico utilizado hasta el momento <b>(m/s)</b> tenemos que <b>x'(o) = 30,55</b>.

<p align="justify">Por otra parte para establecer las constantes <b>C ,K, M </b>adecuadas conociendo que dentro de la gama de opciones de autopartes podemos elegir un elemento elástico cuya rigidez puede ubicarse entre los valores <b>K = 110 000 N/m a 150 000 N/m</b>. , un amortiguador cuya constante de amortiguamiento puede ubicarse entre los valores <b>c = 5 000 N m/seg a 11000 N m/seg</b> y que ,dependiendo del tipo de chasis que se determine para transportar la carga, el peso del mismo varía y se nos dice que las opciones disponibles combinado con el peso de la carga a transportar, el peso completo puede ubicarse entre los  valores de <b>5500 kg y 8000 kg</b>.

<p align="justify">Entonces ante tales indeterminaciones comenzaremos eligiendo un peso medio en el rango propuesto conjeturando que el vehículo será de tamaño regular y que las autopartes para soportar una carga mayor requerirán mayores costos, habrá menos disponibilidad de los mismos o pueden incluso no existir. ```python

```python
# Como notamos que esto sucederá para el mayor K y mayor C de nuestras opciones disponibles,
# entonces resolviendo con el software tenemos:


C_max=11000  #mayor posibilidad de constante de amortiguamiento
K_max=150000 #mayor posibilidad de constante de elasticidad.
#K=185000 #Valor promedio
#C = 8000 # Valor promedio
#M=1687.5 #Valor promedio


W = sqrt((K/M)-((C/(2*M))**2))/(2*pi) #f fundamental del sistema no amortiguado en hertz
Wo = sqrt((K/M))/(2*pi) #f natural del sistema sin amortiguamiento en hertz

error= (Wo.subs(K ,K_max).subs(M,1687.5)) - (W.subs(C,C_max).subs(K ,K_max).subs(M,1687.5))
print(float(error))
```

<p align="justify">Entonces ahora teniendo todos los datos necesarios (constantes y condiciones iniciales) procederemos a resolver la Ecuación Diferencial y así encontrar la función <b>x(t)</b> correspondiente al desplazamiento de la masa del vehículo mientras se mueve por la carretera de peor condición.

```python
#Introducimos las constantes calculadas
EDO_no_homogenea_con_constantes = EDO_no_homogenea.subs(C,7200).subs(K,130000).subs(M,1687.5)

# Definimos condiciones iniciales de posicion y velocidad.
CIni = {x(0): 0 , x(t).diff(t).subs(t,0):0}
# Con estas condiciones establecemos que nuestro vehículo comenzará a moverse desde la posición de equilibrio
# (valor 0 de nuestro sistema de referencia) y con una velocidad vertical igual a 0 ya que consideramos que
# nuestro movil al ingresar al tramo de carretera venia desplazándose solamente de manera horizontal
# ,es decir sin ningún movimiento vertical previo.

x_t= dsolve(EDO_no_homogenea_con_constantes, ics=CIni)
# Resolvemos la EDO.
display(x_t)

plot(x_t.rhs, (t, 0, 4),ylabel = 'x(t)' , xlabel ='t')
# Graficamos la solución en el intervalo de tiempo [0,4].

print("\nAmpliamos el tiempo de visión")

plot(x_t.rhs, (t, 3, 8),ylabel = 'x(t)' , xlabel ='t')
# Graficamos la solución en el intervalo de tiempo [4,8].
```

<p align="justify">• Que la velocidad vertical de la rueda influye en la fuerza del amortiguador, y su signo negativo nos indica que la disminuye.

<p align="justify">**De la segunda ecuación notamos:**

<p align="justify">• La presencia de dos fuerzas amortiguadoras y dos fuerzas elásticas, lo que es sensato ya que la masa de la rueda, además de tener su propio comportamiento de suspensión, también está conectada de manera directa con el amortiguador y el elemento elástico del sistema superior.

<p align="justify">• La presencia de una función $u$ que depende únicamente del tiempo, que influye en la fuerza elástica ejercida por la llanta y su velocidad (derivada primera), y afecta a la fuerza de amortiguamiento, en ambos casos la interacción es negativa, es decir, atenúa dichas fuerzas.

<p align="justify">• La presencia de las mismas fuerzas $F_C$ y $F_K$ del sistema de suspensión, pero con signos opuestos a los presentes en la primera ecuación, por lo que afectan en sentido contrario a la rueda que a lo que lo hacen con la masa del móvil (chasis + carga).



A su vez, este sistema de ecuaciones diferenciales admite la clasificación de sistema ordinario lineal de segundo orden no homogéneo. Para poder encontrar las funciones soluciones, expresaremos el sistema dado como un sistema de ecuaciones de primer orden.

Para ello introduciremos dos nuevas funciones una <b>V_1  = X_1 '</b> y <b>V_2 = x_2 '</b> , y reescribiendo el sistema teniendolas en cuenta, nos queda el siguiente sistema de cuatro ecuaciones y cuatro incognitas:


sis2.png

<p align="justify">Para resolver el sistema, tendremos que asignar valores a las constantes. Con el fin de mantener el análisis del modelo previo, los valores de M, C y K serán los mismos que en el sistema que no consideraba a la rueda como parte del mismo. En cuanto a las demás constantes, se nos ha proporcionado la información de que la constante de amortiguamiento de la rueda <b>(C_R)</b> puede ubicarse entre los valores de <b>200 000 N·m/s</b> a <b>400 000 N·m/s</b>, y su elemento elástico <b>(K_R)</b> entre los valores de <b>1000 N/m</b> a <b>4000 kN/m</b>.
<p align="justify">Para ello, seguiremos la misma lógica que antes y seleccionaremos los valores medios del rango disponible. Así, <b>C_R</b> = 300 000 N·m/s y <b>K_R</b> = 2500 N/m. En cuanto a la masa de la rueda, por no contar con información sobre ella, asumiremos que será de <b>30 kg</b>, valor que se obtuvo de una búsqueda en la web sobre el peso de las ruedas de los camiones de carga más pequeños.
<p align="justify">Por otra parte, la función <b>u(t)</b> que aparece en el sistema ha sido proporcionada y la podemos asociar a una perturbación externa, lo que convierte nuestro sistema en uno no homogéneo. Esta perturbación podría provenir, de manera similar al sistema analizado previamente, del movimiento que le imparte cierta superficie por la que se mueve el transporte.

```python
import numpy as np #importamos la libreria de numpy

def sistema(X, t):

    X1,V1, X2, V2 = X #Se define el vector X de incognitas, en este caso con cuatro elementos X1, V2 , X2 , V2


    #Nuevos coeficientes
    C_R=300000
    K_R=2500
    M_R=30

    #Coeficientes analogos a los del sitema anterior
    C=7200
    K=130000
    M=1687.5

    # Función dada como dato
    U= exp(-2*t)-exp(-4*t)
    U_prima=(-2*exp(-2*t))+(4*exp(-4*t))


    dxdt = [ V1 , (1/M)* ( -C*(V1-V2) -K*(X1 - X2)) , V2   , (1/M_R) * ( -K_R*(X2-U) - C_R*(V2 - U_prima) + C*(V1-V2) + K*(X1 - X2) )]
    return dxdt

#definimos condiciones iniciales
CI = [0 , 0 ,0,0]
# Con estas condiciones establecemos que tanto el chasis con la carga y la rueda comenzarán a moverse desde
# la posición de equilibrio (valor 0 de nuestro sistema de referencia) y con una velocidad vertical igual
# a 0 ya que consideramos que nuestro movil al producirse la perturbación externa venia desplazándose
# solamente de manera horizontal, es decir sin ningún movimiento vertical previo.


# definimos el lapso de tiempo para el cual vamos a resolver el sistema, en este caso entre 0 y 4,
#con un paso tal que se tengamos 200 muestras entre 0 y 4.
t = np.linspace(0, 4, 200)

from scipy.integrate import odeint #importamos la función odeint

sol = odeint(sistema, CI, t)
#pasamos como parametro el sistema, las CI y el tiempo
# y se obtiene la solución del sistema


```

```python
import matplotlib.pyplot as plt

plt.plot(t, sol[:, 0], 'b', label='x1')

#la expresión sol[:, 0] permite graficar la primera coordenada (la cero) de todos los elementos del vector sol,
#es decir, nuestro X1

plt.plot(t, sol[:, 2], 'g', label='x2')

#la expresión sol[:, 2] permite graficar la tercer coordenada (la dos) de todos los elementos del vector sol,
#es decir nuestro X2

plt.xlabel('tiempo (s)')
plt.ylabel('posición (m)')
plt.grid()
plt.show()
```

<p align="justify">En la gráfica podemos ver a la función correspondiente a <b>x1(t)</b> en color azul, que representa el movimiento vertical del chasis más la carga del vehículo a lo largo del tiempo, y a <b>x2(t)</b> en verde, que corresponde al movimiento vertical de la rueda a lo largo del tiempo. En ambos casos, notamos que existe un intervalo de elevación y otro de descenso. <p align="justify">Como ambas funciones comienzan desde la posición de equilibrio, es decir, con una posición de valor 0 en nuestro sistema de referencia, y a su vez con velocidad vertical inicial nula, entonces debido a que no hay otro factor externo que cause movimiento en nuestro sistema más que la <b>u(t)</b>, será ésta función la responsable de la perturbación del mismo. <p align="justify">Para entender mejor el comportamiento que toman las funciones <b>x1</b> y <b>x2</b> en nuestro sistema, analizaremos primero la función <b>u(t)</b> y para ello, la graficaremos.

```python
t = symbols('t')
u = exp(-2*t)-exp(-4*t)
plot(u, (t, 0, 4),ylabel = 'Posición (m)' , xlabel ='Tiempo (s)')
```

<p align="justify">Si comparamos la gráfica de <b>u(t)</b> con la correspondiente a nuestras soluciones, notamos cierta similitud, ya que también aparece la elevación y el posterior descenso. Además, hay una visible semejanza con la gráfica de <b>x2(t)</b> de nuestro sistema, lo que es razonable, ya que en la ecuación correspondiente aparecía de manera explícita una relación. Por otro lado, notamos que la gráfica de <b>x1(t)</b> presenta ciertas oscilaciones visibles que la diferencian.

<p align="justify">Ahora bien, si interpretamos las gráficas en el contexto físico de nuestro problema, considerando a <b>u(t)</b> como una perturbación proveniente del estado del camino por el que se desplaza nuestro vehículo de carga, entendemos que es lógico pensar que, como la rueda está en contacto directo con la superficie y debido a los componentes propios de la misma (<b>M_R</b>, <b>K_R</b> , <b>C_R</b>), su comportamiento será más similar a la perturbación que el correspondiente al desplazamiento de la masa del vehículo (chasis + carga). Esto se debe a que, para que la influencia de esta última perturbación llegue a este último, deberá pasar a través de dos sistemas de amortiguamiento que modificarán su efecto.

<p align="justify">En cuanto a las oscilaciones presentes en <b>x1(t)</b> vemos que estas harán que el descenso de la masa de nuestro móvil no sea de manera uniforme sino que lo haga de una manera más bien intermitente y asociamos esto al efecto producido por los sistemas de amortiguamiento y que aporta confort en el traslado.

<p align="justify"> También estimamos de manera visual  que cuando ambos componentes (rueda y chasis ) lleguen a la mayor elevación, entre  ellos habrá una separación de aproximadamente  de casi 0,10 metros  (10 centímetros) respecto a la posición relativa que tenían en el inicio  una de otra y que será esta la mayor separación en lo que dura el efecto de la perturbación. Esto deberá ser tenido a la hora de la elección de los componentes  ya que los materiales deben resistir dicho estiramiento.

<p align="justify"> Este valor máximo  de la funciones, destacado anteriormente, también representa un cambio brusco en la dirección vertical del vehículo y por tanto de su contenido, lo que vuelve a destacar la importancia de que las cargas se encuentran sujetas adecuadamente en su interior con el fin de evitar accidentes que comprometan la integridad de las vacunas.

<p align="justify">En cuanto la duración del efecto de la perturbación notamos que no es permanente sino que luego de cierto tiempo tiende a desaparecer y algebraicamente es comprobable en la expresión de <b>u(t)</b> que contiene exponenciales negativas. Y si bien matemáticamente el tiempo para que esto ocurra es infinito sabemos que físicamente es posible determinar aproximadamente un tiempo donde la perturbación haya desparecido.

<p align="justify">Considerando entonces al efecto que produce <b>u(t)</b> en el sistema como el estado transitorio de nuestras soluciones y habiendo establecido que será la única perturbación presente, es posible deducir que el movimiento vertical (funciones <b>x1</b> y <b>x2</b> ) también van a tender desaparecer (se acercan al valor 0) por efecto del amortiguamiento que disipa la energia. Interpretamos entonces a la posición de equilibrio (valor 0 en nuestro sistema de referencia) como el estado estacionario al que tenderán nuestras soluciones con el transcurso del tiempo.



<p align="justify">Para determinar cuánto tarda nuestro sistema en alcanzar el equilibrio utilizamos la gráfica y hacemos una estimación visual del tiempo en que las soluciones parecen acercarse bastante al valor 0 comportándose las funciones de manera casi rectilíneas y este parece ser un tiempo aproximado de 3.5 segundos.

<p align="justify">Físicamente significa que tanto nuestro vehículo (chasis + carga) como la rueda dejaran de tener movimiento vertical.
