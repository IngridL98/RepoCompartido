```python
#Importamos las librerias con las funciones que utilizaremos
from sympy import symbols, Function, Derivative,Eq ,plot,plot_implicit, solve, dsolve, simplify, exp, sqrt,pi, cos, simplify,sin,evaluate,And, diff
```

```python
#Declaramos las funciones y las variables simb칩licas

t = symbols('t')  #define la variable independiente que ser치 el tiempo
x = Function('x') #define una funci칩n movimiento respecto del tiempo
M = symbols('M')  #define la variable correspondiente a la masa
K = symbols('K')  #define la variable correspondiente a la constante de elasticidad
C = symbols('C')  #define la variable correspondiente a la constante de amortiguamiento
```

```python
#Definimos la EDO homogenea del sistema
EDO = Eq( M * Derivative(x(t),t,t) + C * Derivative(x(t),t) + K * x(t),0)
```

```python
EDO
```

```python
#Resolvemos la Ecuacion Diferencial
EDO_sol = dsolve(EDO)
```

## An치lisis del confort y desplazamiento sobre una superficie

<p align="justify">Uno de los requerimientos del amortiguamiento que se nos ha solicitado para el traslado es poder garantizar un grado de confort para la carga, es por esto que tomaremos como referencia estudio que determina que el confort se alcanza cuando la <b>frecuencia fundamental (de la oscilaci칩n libre amortiguada) del sistema (w)</b> se encuentra entre los 1  3 Hz y la <b>relaci칩n de amortiguamiento del sistema (E)</b> est치 entre los 0,24 y 0,26. Tener presente que esta relaci칩n est치 dada por la constante de amortiguamiento del sistema y la constante de amortiguamiento cr칤tica del mismo <b>E = C/Cc</b>.

<p align="justify">Habiendo ya calculado <b>w</b> anteriormente, ahora nos falta obtener una expresi칩n para <b>E</b> y para ello debemos calcular el coeficiente de amortiguamiento cr칤tico del sistema, es decir el valor de <b>C</b> que convierte a nuestro sistema en uno cr칤ticamente amortiguado, y algebraicamente lo obtenemos al hallar la expresi칩n para dicho coeficiente que logra hacer 0 el discriminante en la obtenci칩n de las ra칤ces.

<p align="justify">El hecho de que dicho discriminante se anule, har치 que las ra칤ces obtenidas sean reales y repetidas y esto terminar치 afectando con un factor <b>t</b> en una de nuestras soluciones linealmente independientes, provocando as칤 el m치ximo amortiguamiento del sistema. Si bien en nuestra situaci칩n no es deseado ni nos interesa que ello ocurra, obtener este valor <b>Cc</b> ser치 necesario para armar la relaci칩n de amortiguamiento que nos influye en la determinaci칩n del confort.

<p align="justify">Usando el software obtenemos:

```python
Cc = symbols('Cc') # Definimos la constante de amortiguamiento cr칤tica
Ec_discr_igual_cero=Eq (C**2 - 4 *M*K, 0) # Igualamos el discriminante a cero
Cc= solve(Ec_discr_igual_cero,C).pop(1) #Encontramos el valor de Cc que satisface la ec.
Cc
```

```python
E = symbols('E') # definimos la constante
E = C / Cc  # calculamos la relacion de amortiguamiento del sistema
E
```

<p align="justify">All칤 observamos que habi칠ndose considerado una posici칩n de equilibrio, que en nuestro sistema de referencia ser치 el valor x(t) = 0, del desplazamiento vertical de la masa, que a su vez sucede cuando la rueda se encuentra en la mitad de la onda, podemos darnos cuenta que dicha superficie mover치 peri칩dicamente al sistema desde un lugar igual a la amplitud  pero con signo negativo hasta una altura igual a la amplitud, habiendo previamente pasado por la posici칩n de equilibrio. Por lo tanto este movimiento deber치 ser considerado ahora en un nuevo modelo.

<p align="justify">Por su parte se nos ha dicho que este tramo ser치 la peor condici칩n de la carretera y que por tanto ser치 importante lograr que las autopartes que elijamos puedan cumplir con los requerimientos de amortiguaci칩n all칤. Este tramo presenta una superficie ondulada con una longitud de onda de <b>6 metros</b> y una amplitud de ondulaci칩n de la superficie de <b>0,035 m</b>, la cual se recorrer치 a una velocidad de <b>110 km/h</b>.

<p align="justify">Con esta informaci칩n es posible para nosotros representar la carretera mediante una funci칩n con comportamiento sinusoidal y en nuestro caso elegiremos la funci칩n <b>seno</b> ya que consideraremos que el comienzo de las irregularidades del camino se producir치 en la parte media de la onda (valor 0 en nuestro sistema de referencia).

<p align="justify">Entonces sabiendo que dicha funci칩n ser치 de la forma <b>A sen (<i>f</i> m )</b> , en donde <b>A</b> es la amplitud (0,035) y  <i><b>f</b></i> ser치 la frecuencia de la funci칩n trigonom칠trica (2 / 6) que nos permitir치 lograr la longitud de onda deseada <b>(sin considerar la velocidad a칰n)</b>, representaremos la posici칩n de la carretera respecto al equilibrio en funci칩n de los metros (<b>m</b>) mediante el software de la siguiente forma:


```python
A = symbols('A')  #define la variable correspondiente a la amplitud
f = symbols('f')  #define la variable correspondiente a la frecuencia
m = symbols('m')  #define la variable independiente metros
A = 0.035 # asignamos el dato de la amplitud
f = ((2*pi)/6) # calculo de la frecuencia que nos da una longitud de onda de 6 metros
carretera = A * sin(f * m) #armamos la funci칩n
display(carretera)
plot(carretera, (m, 0, 20),ylabel = 'posici칩n (m)' , xlabel ='longitud (m)')
# Graficamos la funcion carreteara en el intervalo [0,20] metros.
```

```python
#Como podemos comprobrar a los 6 metros la carrertera habr치 completado el periodo volviendo a la posici칩n de
#equilibrio 0, es decir su longitud de onda es de 6 metros.

carretera_6m = carretera.subs(m, 6)
print(carretera_6m)
```

<p align="justify">Si consideremos al sistema como lo hab칤amos modelizada anteriormente mediante la ecuaci칩n diferencial de segundo orden <b>M * x''(t) + C* x'(t) + K* x(t) = 0</b> que obtuvimos a partir de la segunda ley de Newton y las fuerzas intervinientes en la situaci칩n y ahora agregamos un nuevo movimiento <b>v(t)</b> que ser치 impartido a la masa (como se visualiza en la siguiente figura), este se acoplar치  al que ten칤amos previamente afectando por tanto al comportamiento de cada parte del sistema de amortiguamiento.

<p align="justify">En consecuencia  se llega a la ecuaci칩n <b>M * x''(t) + C* ( x'(t) - v'(t) ) + K*( x(t) - v(t) ) = 0</b> y despejando los t칠rminos hallamos una expresi칩n no homog칠nea de la forma <b> M * x''(t) + C* x'(t)  + K* x(t) = C* v'(t) + K*v(t) </b>.

image.png

<p align="justify"> Y habiendo ya modelado una funci칩n representativa de la carretera ser치 칠sta la que tendremos en cuenta para armar nuestra <b> v(t)</b> pero <b> modificandole la frecuencia para lograr la velocidad de 30,55 m/s</b> que ser치 la que tendr치 el movil al recorrerla. Por tanto la funci칩n que representa el movimiento impartido en funci칩n del tiempo por la carretera sobre nuestro veh칤culo ser치:
</br><b>v(t)= 0,035 * sen ( ( (2 pi * 30.55)) / 6 ) * t ) </b>.

```python
#Definimos la funci칩n V(t)
v = symbols('v')
v = A * sin(30.55*f * t)
display(v)
plot(v, (t, 0, 1),ylabel = 'posici칩n (m)' , xlabel ='tiempo (s)')

# Graficamos la funcion en el intervalo [0,1] y se puede observar que ahora la longitud de 6 metros
# se recorrer치 30.55/6 veces en un segundo, es decir cumple con la velocidad indicada.
```

```python
v_dif= diff(v, t)

#Definimos la EDO no homogenea del sistema
EDO_no_homogenea = Eq(C*Derivative(x(t), t) + K*x(t) + M*Derivative(x(t), (t, 2)),C*v_dif + K*v)
display(EDO_no_homogenea)
```

<p align="justify">Entonces estableciendo que la posici칩n inicial <b>x(to)=Xo</b> ser치 con el tiempo cero y este ser치 el tiempo en donde las ruedas del m칩vil hagan contacto con la carretera modelada, en ese instante por la funci칩n elegida estaremos a la altura <b>0</b> y estableceremos que nuestro elemento el치stico no estar치 estirado, entonces tenemos <b>x(0)= 0</b> . Por otro lado fue dada como dato la velocidad a la cual se recorre el tramo y es de <b>110 km/h</b> , por tanto haciendo la conversi칩n al sistema m칠trico utilizado hasta el momento <b>(m/s)</b> tenemos que <b>x'(o) = 30,55</b>.

<p align="justify">Por otra parte para establecer las constantes <b>C ,K, M </b>adecuadas conociendo que dentro de la gama de opciones de autopartes podemos elegir un elemento el치stico cuya rigidez puede ubicarse entre los valores <b>K = 110 000 N/m a 150 000 N/m</b>. , un amortiguador cuya constante de amortiguamiento puede ubicarse entre los valores <b>c = 5 000 N m/seg a 11000 N m/seg</b> y que ,dependiendo del tipo de chasis que se determine para transportar la carga, el peso del mismo var칤a y se nos dice que las opciones disponibles combinado con el peso de la carga a transportar, el peso completo puede ubicarse entre los  valores de <b>5500 kg y 8000 kg</b>.

<p align="justify">Para poder tener una vista m치s clara sobre de las opciones que podemos elegir de estas dos constantes trataremos de llevar a una en funci칩n de la otra. ```python

```python
# Como notamos que esto suceder치 para el mayor K y mayor C de nuestras opciones disponibles,
# entonces resolviendo con el software tenemos:


C_max=11000  #mayor posibilidad de constante de amortiguamiento
K_max=150000 #mayor posibilidad de constante de elasticidad.
#K=185000 #Valor promedio
#C = 8000 # Valor promedio
#M=1687.5 #Valor promedio


W = sqrt((K/M)-((C/(2*M))**2))/(2*pi) #f fundamental del sistema no amortiguado en hertz
Wo = sqrt((K/M))/(2*pi) #f natural del sistema sin amortiguamiento en hertz

error= (Wo.subs(K ,K_max).subs(M,1687.5)) - (W.subs(C,C_max).subs(K ,K_max).subs(M,1687.5))
print(float(error))
```

<p align="justify">Entonces ahora teniendo todos los datos necesarios (constantes y condiciones iniciales) procederemos a resolver la Ecuaci칩n Diferencial y as칤 encontrar la funci칩n <b>x(t)</b> correspondiente al desplazamiento de la masa del veh칤culo mientras se mueve por la carretera de peor condici칩n.

```python
#Introducimos las constantes calculadas
EDO_no_homogenea_con_constantes = EDO_no_homogenea.subs(C,7200).subs(K,130000).subs(M,1687.5)

# Definimos condiciones iniciales de posicion y velocidad.
CIni = {x(0): 0 , x(t).diff(t).subs(t,0):0}
# Con estas condiciones establecemos que nuestro veh칤culo comenzar치 a moverse desde la posici칩n de equilibrio
# (valor 0 de nuestro sistema de referencia) y con una velocidad vertical igual a 0 ya que consideramos que
# nuestro movil al ingresar al tramo de carretera venia desplaz치ndose solamente de manera horizontal
# ,es decir sin ning칰n movimiento vertical previo.

x_t= dsolve(EDO_no_homogenea_con_constantes, ics=CIni)
# Resolvemos la EDO.
display(x_t)

plot(x_t.rhs, (t, 0, 4),ylabel = 'x(t)' , xlabel ='t')
# Graficamos la soluci칩n en el intervalo de tiempo [0,4].

print("\nAmpliamos el tiempo de visi칩n")

plot(x_t.rhs, (t, 3, 8),ylabel = 'x(t)' , xlabel ='t')
# Graficamos la soluci칩n en el intervalo de tiempo [4,8].
```

A su vez, este sistema de ecuaciones diferenciales admite la clasificaci칩n de sistema ordinario lineal de segundo orden no homog칠neo. Para poder encontrar las funciones soluciones, expresaremos el sistema dado como un sistema de ecuaciones de primer orden.

Para ello introduciremos dos nuevas funciones una <b>V_1  = X_1 '</b> y <b>V_2 = x_2 '</b> , y reescribiendo el sistema teniendolas en cuenta, nos queda el siguiente sistema de cuatro ecuaciones y cuatro incognitas:

<p align="justify">Para resolver el sistema, tendremos que asignar valores a las constantes. Con el fin de mantener el an치lisis del modelo previo, los valores de M, C y K ser치n los mismos que en el sistema que no consideraba a la rueda como parte del mismo. En cuanto a las dem치s constantes, se nos ha proporcionado la informaci칩n de que la constante de amortiguamiento de la rueda <b>(C_R)</b> puede ubicarse entre los valores de <b>200 000 N췅m/s</b> a <b>400 000 N췅m/s</b>, y su elemento el치stico <b>(K_R)</b> entre los valores de <b>1000 N/m</b> a <b>4000 kN/m</b>.
<p align="justify">Para ello, seguiremos la misma l칩gica que antes y seleccionaremos los valores medios del rango disponible. As칤, <b>C_R</b> = 300 000 N췅m/s y <b>K_R</b> = 2500 N/m. En cuanto a la masa de la rueda, por no contar con informaci칩n sobre ella, asumiremos que ser치 de <b>30 kg</b>, valor que se obtuvo de una b칰squeda en la web sobre el peso de las ruedas de los camiones de carga m치s peque침os.
<p align="justify">Por otra parte, la funci칩n <b>u(t)</b> que aparece en el sistema ha sido proporcionada y la podemos asociar a una perturbaci칩n externa, lo que convierte nuestro sistema en uno no homog칠neo. Esta perturbaci칩n podr칤a provenir, de manera similar al sistema analizado previamente, del movimiento que le imparte cierta superficie por la que se mueve el transporte.

```python
import numpy as np #importamos la libreria de numpy

def sistema(X, t):

    X1,V1, X2, V2 = X #Se define el vector X de incognitas, en este caso con cuatro elementos X1, V2 , X2 , V2


    #Nuevos coeficientes
    C_R=300000
    K_R=2500
    M_R=30

    #Coeficientes analogos a los del sitema anterior
    C=7200
    K=130000
    M=1687.5

    # Funci칩n dada como dato
    U= exp(-2*t)-exp(-4*t)
    U_prima=(-2*exp(-2*t))+(4*exp(-4*t))


    dxdt = [ V1 , (1/M)* ( -C*(V1-V2) -K*(X1 - X2)) , V2   , (1/M_R) * ( -K_R*(X2-U) - C_R*(V2 - U_prima) + C*(V1-V2) + K*(X1 - X2) )]
    return dxdt

#definimos condiciones iniciales
CI = [0 , 0 ,0,0]
# Con estas condiciones establecemos que tanto el chasis con la carga y la rueda comenzar치n a moverse desde
# la posici칩n de equilibrio (valor 0 de nuestro sistema de referencia) y con una velocidad vertical igual
# a 0 ya que consideramos que nuestro movil al producirse la perturbaci칩n externa venia desplaz치ndose
# solamente de manera horizontal, es decir sin ning칰n movimiento vertical previo.


# definimos el lapso de tiempo para el cual vamos a resolver el sistema, en este caso entre 0 y 4,
#con un paso tal que se tengamos 200 muestras entre 0 y 4.
t = np.linspace(0, 4, 200)

from scipy.integrate import odeint #importamos la funci칩n odeint

sol = odeint(sistema, CI, t)
#pasamos como parametro el sistema, las CI y el tiempo
# y se obtiene la soluci칩n del sistema


```

```python
import matplotlib.pyplot as plt

plt.plot(t, sol[:, 0], 'b', label='x1')

#la expresi칩n sol[:, 0] permite graficar la primera coordenada (la cero) de todos los elementos del vector sol,
#es decir, nuestro X1

plt.plot(t, sol[:, 2], 'g', label='x2')

#la expresi칩n sol[:, 2] permite graficar la tercer coordenada (la dos) de todos los elementos del vector sol,
#es decir nuestro X2

plt.xlabel('tiempo (s)')
plt.ylabel('posici칩n (m)')
plt.grid()
plt.show()
```

En ambos casos, notamos que existe un intervalo de elevaci칩n y otro de descenso. <p align="justify">Como ambas funciones comienzan desde la posici칩n de equilibrio, es decir, con una posici칩n de valor 0 en nuestro sistema de referencia, y a su vez con velocidad vertical inicial nula, entonces debido a que no hay otro factor externo que cause movimiento en nuestro sistema m치s que la <b>u(t)</b>, ser치 칠sta funci칩n la responsable de la perturbaci칩n del mismo. <p align="justify">Para entender mejor el comportamiento que toman las funciones <b>x1</b> y <b>x2</b> en nuestro sistema, analizaremos primero la funci칩n <b>u(t)</b> y para ello, la graficaremos.
